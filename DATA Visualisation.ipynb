```python
fig, ax = plt.subplots(figsize=(10,6))
brand = df.groupby('Battery(mAh)')['Price(€)'].mean().reset_index()
brand.sort_values(by='Price(€)', ascending=True, inplace=True)
plt.scatter(data=brand, x='Battery(mAh)', y='Price(€)')
plt.xlabel("Battery(mAh)")
plt.ylabel("Price(€)")
```

```python
brand = df.groupby('Marque')['Price(€)'].mean().reset_index()
brand.sort_values(by='Price(€)', ascending=False, inplace=True)
plt.figure(figsize=(12,16))
sns.barplot(data=brand[brand['Price(€)'] < 0.1e5], y='Marque', x='Price(€)')
plt.ylabel(None)
plt.xlabel('Suggested Retail Price')
plt.title("Average Price of smartphone per Brand")
plt.show()
```
